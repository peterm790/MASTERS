```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy
from matplotlib.cm import get_cmap
import matplotlib
```

```python
#daily
U = xr.open_dataset('/terra/data/cmip5/global/historical/BNU-ESM/r1i1p1/3hr/native/uas_3hr_BNU-ESM_historical_r1i1p1_195001010000-200512312100.nc')
V = xr.open_dataset('/terra/data/cmip5/global/historical/BNU-ESM/r1i1p1/3hr/native/vas_3hr_BNU-ESM_historical_r1i1p1_195001010000-200512312100.nc')
```

```python
U = U.sel(time=slice('2000-06-01','2000-08-30'))
V = V.sel(time=slice('2000-06-01','2000-08-30'))
```

```python
V
```

```python
V = V.rename({'lon':'longitude'})
V = V.rename({'lat':'latitude'})
U = U.rename({'lon':'longitude'})
U = U.rename({'lat':'latitude'})
U = U.rename({'uas':'uwnd'})
V = V.rename({'vas':'vwnd'})
U = U.sel(longitude = slice(-40,30))
U = U.sel(latitude = slice(-75,-15))
V = V.sel(longitude = slice(-40,30))
V = V.sel(latitude = slice(-75,-15))
```

```python
ws = np.sqrt(np.square(U.uwnd) + np.square(V.vwnd))
Dir=np.mod(180+np.rad2deg(np.arctan2(U.uwnd, V.vwnd)),360)
```

```python
front = xr.open_zarr('FRONT/BNU-ESM_front.zarr')
```

```python
front = front.sel(time=slice('2000-06-01','2000-08-30'))
```

```python
front = front.front/front.front
front = front.fillna(0)
front=front.compute()
```

```python
front.longitude
```

```python
for i in range(len(front.time)):
    ax = plt.axes(projection=ccrs.PlateCarree())
    plt.rcParams['hatch.linewidth']=1
    plt.rcParams['hatch.color']='black'
    ax.coastlines('50m', linewidth=0.8)
    Q = ax.streamplot(U.longitude, U.latitude, U.uwnd.values[i], V.vwnd.values[i],transform = ccrs.PlateCarree(),density = 3, zorder=0)
    W= ax.contourf(U.longitude, U.latitude, ws[i],transform = ccrs.PlateCarree(),alpha=0.6,zorder=1)
    F =ax.contour(U.longitude,U.latitude,front[i],colors='black',levels = [0.99,1.01], transform = ccrs.PlateCarree(),alpha=1,zorder=2)
    F =ax.contourf(U.longitude,U.latitude,front[i],levels = [0.99,1.01], hatches=["x"], transform = ccrs.PlateCarree(),alpha=0,zorder=3)
    ax.set_extent([0, 30, -15, -75], ccrs.PlateCarree())
    plt.title('ERA5 Front Detection - day\n'+str(U.time.values[i]))
    plt.savefig('Front_plot/'+str(i)+'togif.png')
    plt.close()
```
